```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
data=pd.read_csv("Data.csv")
data
```

```python
#scatter plot for the given data
plt.scatter(data['SAT'],data['GPA'])
plt.show()
```

```python
#heatmap for the dataset with readings in it
sns.heatmap(data.corr(),annot=True)
```

```python
from sklearn import linear_model
regress = linear_model.LinearRegression()

train_x = np.asanyarray(data[['SAT']])
train_y = np.asanyarray(data[['GPA']])
#print(train_x)
#print(train_y)

regress.fit (train_x,train_y)
# The coefficients
print ('Coefficients: ', regress.coef_)
print ('Intercept: ',regress.intercept_)

regress1 = linear_model.LinearRegression()
train_x1 = np.asanyarray(data[['SAT']])
train_y1 = np.asanyarray(data[['GPA']])
#print(train_x)
#print(train_y)

regress1.fit (train_x1,train_y1)
# The coefficients
print ('Coefficients: ', regress1.coef_)
print ('Intercept: ',regress1.intercept_)
```

```python
plt.scatter(data.SAT, data.GPA,  color='blue',label="Actual")
plt.plot(train_x, regress.coef_[0][0]*train_x + regress.intercept_[0], '-r',label="Predicted")
plt.title("Placement plot")
plt.legend()
plt.xlabel("SAT")
plt.ylabel("GPA")
plt.show()
```

```python
y_predicted = regress.predict(train_x)
for i in range(0,len(train_x)):
    print(train_y[i],y_predicted[i])
data['Predicted'] = y_predicted
print(y_predicted)
print(data.head())
```

```python
#finding new y values with y=mx+c with coefficent and intercept
y_predicted = regress.predict([[1700]])
print(y_predicted)
```

```python
from sklearn import metrics
print('Mean Absolute Error:', metrics.mean_absolute_error(train_y, y_predicted))
print('Mean Squared Error:', metrics.mean_squared_error(train_y, y_predicted))
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(train_y, y_predicted)))
```
